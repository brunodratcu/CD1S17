### Melhoria das condições sanitárias pelo aumento do PIB influenciam na mortalidade infantil?

```python
%matplotlib inline

import os
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import pandas as pd
import numpy as np

import statsmodels.api as sm1 
import statsmodels.formula.api as sm  

print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
dados1 = pd.read_excel('indicator gapminder under5mortality.xlsx',index_col=0)
dados2 = pd.read_excel('GDPpercapitaconstant2000US.xlsx',index_col=0)
dados3 = pd.read_excel('Indicator_Improved sanitation total percent.xlsx',index_col=0)
```

```python
dados = pd.DataFrame()
dados["Mortalidade_Infantil"] = dados1[2010]
dados["PIB"] = dados2["2010"]
dados["Condições_sanitárias"] = dados3["2010"]
```

```python
dados = dados.dropna()
#dados["Mortalidade_Infantil"] = sorted(dados["Mortalidade_Infantil"])
#dados["PIB"] = sorted(dados["PIB"])
#dados["Condições_sanitárias"] = sorted(dados["Condições_sanitárias"])
dados.describe()
```

Diante da análise é possivel entender a relação entre o índice de mortalidade infantil e as variáveis explicativas. Observando os gráficos de dispersão a baixo, que tem como variaveis explicativas a melhora das condições sanitárias e PIB per capita, apresentam uma disposição dos pontos próxima de linear que pode ser comprovado pelos fortes índices de correlação com mortalidade infantil.

```python
fig = plt.figure(figsize=(10,5))

plt.subplot(121)
plt.scatter(dados["PIB"], dados["Mortalidade_Infantil"], alpha=0.8, color="green")
plt.xlabel("PIB per capita")
plt.ylabel("Mortalidade Infantil")

plt.subplot(122)
plt.scatter(dados["Condições_sanitárias"], dados["Mortalidade_Infantil"], alpha=0.9, color="yellow")
plt.xlabel("Melhora das condições sanitárias")
plt.ylabel("Mortalidade Infantil")

plt.tight_layout()
```

```python
dados.corr(method='pearson')
```

A partir da análise das correlações obtidas para as variáveis explicativas, teve uma correlações altas para PIB e condições sanitárias 0.96 e 0.71 respectivamente.

```python
varis = sm.ols(formula='Mortalidade_Infantil ~ Condições_sanitárias + PIB', data = dados)
fit = varis.fit()
fit.summary()

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111, projection='3d')

x_surf = np.arange(-20, 120, 2)               
y_surf = np.arange(-20, 120, 0.5)
x_surf, y_surf = np.meshgrid(x_surf, y_surf)

exog = pd.core.frame.DataFrame({'PIB': x_surf.ravel(), 'Condições_sanitárias': y_surf.ravel()})
out = fit.predict(exog = exog)
ax.plot_surface(x_surf, y_surf,
                out.reshape(x_surf.shape),
                rstride=1,
                cstride=1,
                color='None',
                alpha = 0.1)

ax.scatter(dados['Condições_sanitárias'], dados['PIB'], dados['Mortalidade_Infantil'],
           c='blue',
           marker='o',
           alpha=0.7)

#ax.set_xlim(0,100)
#ax.set_ylim(0,10)
ax.set_xlabel('PIB')
ax.set_ylabel('Percentual em melhoras de condições sanitárias')
ax.set_zlabel('Mortalidade Infantil')
```
